```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('/content/netflix_titles.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.fillna('Unknown', inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
print(df.columns)
```

```python
df['rating'] = df['rating'].str.strip().str.lower()
```

```python
df['date_addded'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df.head(3)
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df['release_year'] = df['release_year'].astype('int')
```

```python
df.describe()
```

```python
df.to_csv('/content/netflix_titles.csv', index=False)
```
